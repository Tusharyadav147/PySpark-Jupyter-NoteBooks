```python
from pyspark.sql import SparkSession
```

```python
if __name__ == "__main__":
  session = SparkSession.builder.appName('firstsql').master('local[2]').getOrCreate()
  readdata = session.read
```

```python
ibmdata = readdata.option('header' , True).option('inferSchema' ,True).csv('IBMdata.csv')
```

```python
ibmdata.printSchema()
```

```python
# displaying the  records is done via show()
# in RDD we used to diplay we use collect()
result1 = ibmdata.select('Gender','Age')
result1.show()
```

```python
result2 = ibmdata.select('Age','DailyRate',ibmdata.Department.like("S%")).show()
```

```python
result4 = ibmdata.select('Gender',ibmdata.JobRole.startswith('Sales')).show()
```

```python
result5 = ibmdata.select('Department' ,ibmdata.Age.between(60,90))
result5.show()
```

```python
result7 = ibmdata.select('Gender' , ibmdata.Department.isin('Sales','Human Resources'))
result7.show()
```

```python
result9 = ibmdata.select('Gender',ibmdata.JobRole.substr(1,4))
result9.show()
```

```python
result9 = ibmdata.filter((ibmdata['Gender']=="Male") & (ibmdata['MaritalStatus'] == 'Married') | (ibmdata['Age']> 50))
result9.show()
```

```python
result10 = ibmdata.filter((ibmdata['Attrition'] == 'Yes') & (ibmdata['Age'] >= 30 ))
result10.show()
```

```python
result11 = ibmdata.filter((ibmdata['Age'] >40) & (ibmdata['MaritalStatus'] == 'Married')).select('Gender' ,'BusinessTravel')
result11.show()
```

```python
conditon1 = ibmdata.filter((ibmdata['DailyRate'] >500) & (ibmdata['MaritalStatus'] == 'Married')).select('Department','JobRole').show()
```

```python
condition2 = ibmdata.filter((ibmdata['EmployeeNumber'] >= 30) |(ibmdata['EmployeeCount'] > 5) ).select('Department' ,ibmdata.Age.between(50,90)).show()
```

```python
condition3 = ibmdata.filter((ibmdata['Gender'] == 'Male') &(ibmdata['Department']=='Sales')).select('DailyRate','Gender' ,'Department','JobRole').show()
```

```python
condition4 = ibmdata.filter((ibmdata['MaritalStatus'] == 'Married') & (ibmdata['Gender']== 'Female') & (ibmdata['DailyRate'] > 150 ) & (ibmdata['Age'] >50)).select('Age','Attrition','JobRole','MaritalStatus').limit(5).show()
```

```python
contition5 = ibmdata.filter((ibmdata['Department'] == 'Sales') & (ibmdata['MaritalStatus'] != 'Married') & (ibmdata['JobRole'].like('S%'))).select('JobRole','Department','MaritalStatus' ,'Age').show()
```
